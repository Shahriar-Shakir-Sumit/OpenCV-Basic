```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
def plt_imshow(title, image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.title("Cat Image")
    plt.grid(True)
    plt.show()
```

```python
args = {"Image":"catImage.jpg"}
```

```python
img = cv2.imread("catImage.jpg")
img.shape
```

```python
imageAddCustomValue = cv2.add(np.uint8([200]), np.uint8([100]))
imageSubtractCustomValue = cv2.subtract(np.uint8([50]), np.uint8([100]))
```

```python
print("Image Maximum Value of 255 {}".format(imageAddCustomValue))
print("Image Minimum Value of 255 {}".format(imageSubtractCustomValue))
```

```python
imageAddCustomValue = np.uint8([200]) + np.uint8([100])
imageSubtractCustomValue = np.uint8([5]) - np.uint8([10])
```

```python
print("Image Maximum Value of 255 {}".format(imageAddCustomValue))
print("Image Minimum Value of 255 {}".format(imageSubtractCustomValue))
```
